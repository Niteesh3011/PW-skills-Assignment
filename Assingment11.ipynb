#### Question 1: What is hypothesis testing in statistics?

Answer ->

Hypothesis testing is a fundamental concept in statistics that allows us to make decisions about a population based on sample data. It's a formal procedure for investigating our ideas about the world using statistics. 


At its core, hypothesis testing is a method of statistical inference. We use it to determine whether there is enough evidence in a sample of data to draw conclusions about a larger population.


The Two Competing Hypotheses
Every hypothesis test involves two competing statements about the population:

The Null Hypothesis (H₀): This is a statement of no effect or no difference. It's the default assumption that we're trying to find evidence against. For example, a null hypothesis might state that a new drug has no effect on a particular illness.


The Alternative Hypothesis (Hₐ or H₁): This is the statement that we are trying to find evidence to support. It's the opposite of the null hypothesis. In the drug example, the alternative hypothesis would be that the new drug does have an effect.

#### Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

Answer ->

##### The Role of the Significance Level in Decision Making

The significance level plays a direct role in deciding the outcome of a hypothesis test by providing a cutoff point. Here's how it works in conjunction with the p-value:

The p-value is a calculated probability from your sample data. It represents the probability of observing your results (or more extreme results) if the null hypothesis were true.

The decision rule is straightforward:

If the p-value ≤ α (significance level), you reject the null hypothesis. This means your results are statistically significant, and you have enough evidence to support the alternative hypothesis.

If the p-value > α (significance level), you fail to reject the null hypothesis. This means your results are not statistically significant, and you don't have enough evidence to support the alternative hypothesis.

#### Question 4: What are Type I and Type II errors? Give examples of each.

Answer ->

In hypothesis testing, we make decisions about a population based on sample data, and sometimes, those decisions are wrong. Type I and Type II errors are the two kinds of mistakes we can make.

A Type I error happens when you reject the null hypothesis when it is actually true. Think of it as a "false positive." The probability of making a Type I error is equal to the significance level, denoted by alpha (α).

A Type II error occurs when you fail to reject the null hypothesis when it is actually false. This is like a "false negative." The probability of making a Type II error is denoted by beta (β).

#### Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.

Answer ->

Use a Z-Test When:

You should use a Z-test if you meet both of these conditions:

Your sample size is large (typically n > 30).

You know the standard deviation of the population (σ).

It's quite rare in practice to know the population standard deviation. This makes the Z-test less commonly used than the T-test.

Example Scenario: A school administrator wants to know if the IQ scores of students in their school differ from the national average. The national IQ scores have a known standard deviation. The administrator takes a sample of 40 students. In this case, a Z-test would be appropriate because the sample size is large, and the population standard deviation is known.

Use a T-Test When:

You should use a T-test if you meet either of these conditions:

Your sample size is small (typically n < 30).

You do not know the standard deviation of the population (σ) and must estimate it from your sample data.

Example Scenario: A researcher wants to test if a new fertilizer increases the height of a specific type of plant. They plant 20 seeds with the new fertilizer. Since they don't know the standard deviation for all plants grown with this new fertilizer (the population), and the sample size is small, they would use a T-test. The standard deviation of the sample is used as an estimate.

#### Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def plot_binomial_distribution(n, p, size=1000):
    """
    Generates data for a binomial distribution and plots its histogram.

    Args:
        n (int): The number of trials in each experiment.
        p (float): The probability of success in each trial (must be between 0 and 1).
        size (int): The total number of experiments to simulate.
    """
    # --- 1. Generate Binomial Distribution Data ---
    # Generate random numbers from a binomial distribution.
    # Each number in the 'data' array represents the total number of successes
    # out of 'n' trials for one of the 'size' experiments.
    data = np.random.binomial(n, p, size)

    # --- 2. Plot the Histogram ---
    # Set a professional style for the plot using seaborn.
    sns.set_style("whitegrid")
    plt.figure(figsize=(10, 6)) # Set the figure size for better readability

    # Create the histogram.
    # 'discrete=True' is ideal for integer data like the binomial distribution.
    # 'stat="frequency"' ensures the y-axis shows the count of occurrences.
    sns.histplot(data, discrete=True, stat="frequency", shrink=0.8)

    # --- 3. Customize and Display the Plot ---
    # Add a title and labels for clarity.
    plt.title(f'Histogram of Binomial Distribution (n={n}, p={p}, size={size})', fontsize=16)
    plt.xlabel('Number of Successes', fontsize=12)
    plt.ylabel('Frequency', fontsize=12)

    # Set the x-axis ticks to be integers representing each possible outcome.
    plt.xticks(range(n + 1))

    # Ensure the layout is tight and clean.
    plt.tight_layout()

    # Display the plot.
    plt.show()

if __name__ == '__main__':
    # --- Parameters for the distribution ---
    # Number of trials
    num_trials = 10
    # Probability of success
    prob_success = 0.5

    # Generate and plot the distribution
    plot_binomial_distribution(n=num_trials, p=prob_success)
```

#### Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results. sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

```python
import numpy as np
from scipy.stats import norm
import math

def perform_z_test(sample_data, pop_mean_hypothesized, pop_std_dev, alpha=0.05):
    """
    Performs a two-tailed Z-test for a population mean.

    Args:
        sample_data (list or np.array): The sample data.
        pop_mean_hypothesized (float): The hypothesized population mean (H₀).
        pop_std_dev (float): The known population standard deviation (σ).
        alpha (float): The significance level for the test.

    Returns:
        None: Prints the results of the test.
    """
    # --- 1. Calculate Sample Statistics ---
    sample_mean = np.mean(sample_data)
    n = len(sample_data)

    print(f"Sample Mean (x̄): {sample_mean:.4f}")
    print(f"Sample Size (n): {n}")
    print("-" * 30)

    # --- 2. Calculate the Z-statistic ---
    # The Z-statistic measures how many standard errors the sample mean is from the
    # hypothesized population mean.
    # Formula: Z = (x̄ - μ) / (σ / √n)
    standard_error = pop_std_dev / math.sqrt(n)
    z_statistic = (sample_mean - pop_mean_hypothesized) / standard_error

    # --- 3. Calculate the P-value ---
    # For a two-tailed test, the p-value is the probability of observing a Z-statistic
    # as extreme as the one calculated, in either tail of the standard normal distribution.
    p_value = 2 * (1 - norm.cdf(abs(z_statistic)))

    # --- 4. Interpret the Results ---
    print(f"Hypothesized Population Mean (μ): {pop_mean_hypothesized}")
    print(f"Population Standard Deviation (σ): {pop_std_dev}")
    print(f"Significance Level (α): {alpha}")
    print("-" * 30)
    print(f"Z-statistic: {z_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")
    print("-" * 30)

    # --- 5. Make a Decision ---
    print("Conclusion:")
    if p_value < alpha:
        print(f"Since the p-value ({p_value:.4f}) is less than the significance level ({alpha}), we reject the null hypothesis.")
        print(f"There is significant evidence to suggest that the true population mean is different from {pop_mean_hypothesized}.")
    else:
        print(f"Since the p-value ({p_value:.4f}) is greater than the significance level ({alpha}), we fail to reject the null hypothesis.")
        print(f"There is not enough significant evidence to suggest that the true population mean is different from {pop_mean_hypothesized}.")

if __name__ == '__main__':
    # The sample dataset provided by the user.
    sample_data = [
        49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
        50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
        50.3, 50.4, 50.0, 49.7, 50.5, 49.9
    ]

    # --- Setup for the Hypothesis Test ---
    # Null Hypothesis (H₀): The true population mean is 50.0.
    # Alternative Hypothesis (Hₐ): The true population mean is not 50.0 (a two-tailed test).
    
    # NOTE: A Z-test assumes the POPULATION standard deviation (σ) is KNOWN.
    # Since it's not provided, we will assume a value for this example.
    # In a real-world scenario with an unknown σ, a T-test would be more appropriate.
    
    hypothesized_mean = 50.0
    known_population_std_dev = 0.5  # Assumed known population standard deviation
    
    perform_z_test(sample_data, hypothesized_mean, known_population_std_dev)
```

#### Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import seaborn as sns

def plot_confidence_interval(mu=100, sigma=15, n=100, confidence_level=0.95):
    """
    Simulates data from a normal distribution, calculates the confidence interval
    for the mean, and plots the results.

    Args:
        mu (float): The true mean of the normal distribution.
        sigma (float): The true standard deviation of the normal distribution.
        n (int): The number of data points to simulate (sample size).
        confidence_level (float): The desired confidence level (e.g., 0.95 for 95%).
    """
    # --- 1. Simulate Data ---
    # Generate 'n' random data points from a normal distribution with a
    # specified mean (mu) and standard deviation (sigma).
    np.random.seed(42) # for reproducibility
    data = np.random.normal(loc=mu, scale=sigma, size=n)

    # --- 2. Calculate Sample Statistics and Confidence Interval ---
    # Calculate the mean and standard deviation of our sample.
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1) # ddof=1 for sample standard deviation

    # Calculate the standard error of the mean (SEM).
    sem = sample_std / np.sqrt(n)

    # Calculate the 95% confidence interval for the mean.
    # We use the t-distribution because we are using the sample standard deviation
    # as an estimate for the unknown population standard deviation.
    ci = st.t.interval(
        confidence=confidence_level,
        df=n - 1,              # Degrees of freedom
        loc=sample_mean,       # Sample mean
        scale=sem              # Standard error of the mean
    )

    # --- 3. Plot the Data and Confidence Interval ---
    sns.set_style("whitegrid")
    plt.figure(figsize=(12, 7))

    # Plot the histogram of the simulated data.
    sns.histplot(data, kde=True, bins=15, label=f'Simulated Data (n={n})', color='skyblue')

    # Add vertical lines to show the means and confidence interval.
    plt.axvline(mu, color='red', linestyle='--', linewidth=2, label=f'True Population Mean (μ={mu})')
    plt.axvline(sample_mean, color='green', linestyle='-', linewidth=2, label=f'Sample Mean (x̄={sample_mean:.2f})')
    plt.axvline(ci[0], color='purple', linestyle=':', linewidth=2, label=f'95% CI Lower Bound ({ci[0]:.2f})')
    plt.axvline(ci[1], color='purple', linestyle=':', linewidth=2, label=f'95% CI Upper Bound ({ci[1]:.2f})')
    
    # Fill the area of the confidence interval for better visualization.
    plt.axvspan(ci[0], ci[1], color='purple', alpha=0.1, label='95% Confidence Interval')


    # --- 4. Customize and Display the Plot ---
    plt.title(f'95% Confidence Interval for the Mean of Simulated Normal Data', fontsize=16)
    plt.xlabel('Value', fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.legend()
    plt.tight_layout()
    
    print(f"Simulated Data Summary:")
    print(f"Sample Size (n): {n}")
    print(f"Sample Mean (x̄): {sample_mean:.4f}")
    print(f"Sample Standard Deviation (s): {sample_std:.4f}")
    print("-" * 40)
    print(f"The {confidence_level*100}% confidence interval for the population mean is: ({ci[0]:.4f}, {ci[1]:.4f})")
    
    plt.show()

if __name__ == '__main__':
    # --- Parameters for the simulation ---
    population_mean = 100
    population_std_dev = 15
    sample_size = 100

    plot_confidence_interval(mu=population_mean, sigma=population_std_dev, n=sample_size)
```
