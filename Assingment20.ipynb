# Theoretical Questions:

Q.1) What is unsupervised learning in the context of machine learning?

Answer->
Unsupervised Learning is a type of machine learning where the algorithm is trained on data that has not been labeled, classified, or categorized. Unlike supervised learning, where the model learns from input-output pairs (ground truth), unsupervised learning deals with finding hidden structures, patterns, or relationships within the input data itself.

Goal: To model the underlying structure or distribution in the data in order to learn more about the data.

Key Techniques:

Clustering: Grouping data points by similarity (e.g., K-Means, DBSCAN).

Dimensionality Reduction: Reducing the number of random variables under consideration (e.g., PCA, t-SNE).

Association Rules: Discovering rules that describe large portions of data (e.g., Apriori algorithm).

Q.2) How does K-Means clustering algorithm work?

Answer ->

K-Means is an iterative algorithm that tries to partition the dataset into 7$K$ pre-defined distinct non-overlapping subgroups (clusters) where each data point belongs to only one group.

The Algorithm Steps:
Initialization: Choose the number of clusters 9$K$.10 Randomly select $K$ data points as the initial centroids (cluster centers).

Assignment: For each data point in the dataset, calculate the distance (usually Euclidean) to each centroid. Assign the data point to the cluster with the nearest centroid.

Update: Recalculate the position of the centroids.

The new centroid of each cluster is the mean (average) of all data points assigned to that cluster.

Convergence: Repeat steps 2 and 3 until:The centroids do not change position (stabilization).The assignments of points do not change.A maximum number of iterations is reached.

Q.3) Explain the concept of a dendrogram in hierarchical clustering.

Answer ->

A dendrogram is a tree-like diagram that records the sequences of merges or splits in hierarchical clustering. It visually represents the arrangement of the clusters produced by the corresponding analysis.

X-axis: Represents the individual data points (or clusters).

Y-axis: Represents the distance (or dissimilarity) between the clusters being merged.

Interpretation: The height at which two branches merge indicates how different the two clusters are. A merge happening at a low height means the clusters were very similar; a merge at a high height means they were quite different.

Cutting the Tree: By drawing a horizontal line across the dendrogram at a specific y-value, you "cut" the tree, which determines the final number of clusters.

Q.5) What are the advantages of DBSCAN over K-Means?

Answer ->

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) offers distinct advantages:

Arbitrary Shapes: DBSCAN can find clusters of arbitrary shapes (e.g., concentric rings, moons).K-Means assumes clusters are spherical/convex.

No $K$ Required: You do not need to specify the number of clusters beforehand.

Noise Handling: DBSCAN explicitly identifies outliers as "noise" and excludes them from clusters.18 K-Means forces every point (even outliers) into a cluster, which can distort the cluster center.

Q.6) When would you use Silhouette Score in clustering?

Answer ->

The Silhouette Score is used to evaluate the quality of clusters created by an algorithm.20 You use it when:

Ground truth is missing: You don't have labeled data to verify if the clusters are correct.

Comparing Models: You want to compare different algorithms (e.g., K-Means vs. Hierarchical).

Tuning Hyperparameters: Specifically, finding the optimal number of clusters (22$K$) in K-Means.23 A higher average silhouette score indicates better-defined clusters.24

Q.7) What are the limitations of Hierarchical Clustering?

Answer ->

Scalability: It has a high time complexity (usually $O(n^3)$ or $O(n^2 \log n)$) and space complexity, making it very slow and memory-intensive for large datasets.

Irreversibility: In Agglomerative clustering, once two clusters are merged, they cannot be separated later.25 If a bad merge decision is made early, it propagates through the rest of the process.

Subjectivity: Choosing where to "cut" the dendrogram to determine the final number of clusters can be subjective.

Q.8) Why is feature scaling important in clustering algorithms like K-Means?

Answer ->

Clustering algorithms often rely on distance measurements (like Euclidean distance) to determine similarity.

Magnitude Bias: If one feature has a large range (e.g., Salary: 20,000–100,000) and another has a small range (e.g., Age: 20–60), the distance calculation will be completely dominated by the Salary feature. The algorithm will effectively ignore Age.

Normalization: Feature scaling (Standardization or Min-Max Scaling) ensures all features contribute equally to the result.

Q.9) ow does DBSCAN identify noise points?

Answer ->

DBSCAN classifies points into three categories: Core, Border, and Noise. A point is identified as Noise (or an outlier) if:

It is not a Core Point (it does not have minPts within its epsilon radius).

It is not reachable from any Core Point (it is not within the epsilon radius of any Core Point).

Essentially, noise points are in low-density regions and do not belong to any cluster.

Q.10)  Define inertia in the context of K-Means.

Answer ->

Inertia (also known as Within-Cluster Sum of Squares - WCSS) measures how internally coherent the clusters are. Mathematically, it is the sum of squared distances of samples to their closest cluster center.

$$\text{Inertia} = \sum_{i=1}^{N} || x_i - \mu_{c_i} ||^2$$

$x_i$: Data point

$\mu_{c_i}$: The centroid of the cluster to which 32$x_i$ belongs.

Goal: K-Means tries to minimize inertia.34 Lower inertia implies denser clusters.

Q.11) What is the elbow method in K-Means clustering?

Answer ->

The Elbow Method is a heuristic used to determine the optimal number of clusters ($K$).

Procedure: You run K-Means for a range of $K$ values (e.g., 1 to 10) and calculate the Inertia for each.

Plot: Plot $K$ (x-axis) vs. Inertia (y-axis).

Interpretation: As $K$ increases, inertia decreases. The "elbow" is the point where the rate of decrease sharply shifts (diminishing returns).42 This point represents the optimal tradeoff between error and number of clusters.

 DBSCAN, density is defined by two parameters:

Q.13) Can hierarchical clustering be used on categorical data?

Answer ->

Yes, but with caution. Standard hierarchical clustering uses Euclidean distance, which is invalid for categorical data (e.g., calculating the distance between "Red" and "Blue" is meaningless). To use it on categorical data:

You must use a suitable distance metric, such as Gower's distance, Hamming distance, or Jaccard distance.

You often need to one-hot encode the data first, depending on the implementation.

Q.14) What does a negative Silhouette Score indicate?

Answer ->

The Silhouette Score ranges from -1 to +1.

+1: The sample is far away from the neighboring clusters (good).

0: The sample is on or very close to the decision boundary between two neighboring clusters.

Negative (< 0): This indicates that the sample has likely been assigned to the wrong cluster. It is closer to a neighboring cluster than to the one it was assigned to.

Q.15) Explain the term "linkage criteria" in hierarchical clustering.

Answer ->

Linkage criteria determine how the distance between two clusters is calculated when deciding which clusters to merge. Common types include:

Single Linkage: Distance between the closest pair of points in the two clusters (can result in long, thin "chain" clusters).

Complete Linkage: Distance between the farthest pair of points in the two clusters (tends to find compact spherical clusters).

Average Linkage: Average distance between all pairs of points in the two clusters.

Ward’s Method: Minimizes the increase in variance (inertia) when merging clusters.

Q.16) Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

Answer ->

K-Means makes a strong assumption that clusters are convex (spherical) and isotropic (similar size/density).

Centroid Logic: K-Means assigns points based purely on distance to the center.

The Issue: If a dense, small cluster is next to a sparse, large cluster, the centroid of the sparse cluster might be far away. Points that conceptually belong to the sparse cluster might be closer to the dense cluster's centroid, leading to misclassification. K-Means does not account for the "spread" (variance) of the cluster, only the center.

Q.17) What are the core parameters in DBSCAN, and how do they influence clustering?

Answer->

Epsilon ($\epsilon$):

Too small: Most data will not be reachable; many points will be labeled as noise.

Too large: Clusters will merge together, potentially putting the whole dataset into one cluster.

MinPts:

Too low: Noise points might form small, meaningless clusters.

Too high: Only extremely dense regions will form clusters; valid but less dense clusters might be classified as noise.

Q.18) How does K-Means++ improve upon standard K-Means initialization?

Answer ->

Standard K-Means initializes centroids completely randomly, which can lead to poor convergence or getting stuck in local optima.

K-Means++ improves this by spreading out the initial centroids:

Choose the first centroid randomly from the data.For every other data point, calculate the distance ($D(x)$) to the nearest existing centroid.Choose the next centroid from the data points with a probability proportional to $D(x)^2$. (Points further away are more likely to be chosen).Repeat until 59$K$ centroids are chosen.60This ensures centroids are far apart, leading to faster convergence and better results.61

Q.19) What is agglomerative clustering?

Answer ->

Agglomerative Clustering is a "bottom-up" approach to hierarchical clustering.

Start: Treat every single data point as its own individual cluster (if you have $N$ points, you start with $N$ clusters).

Iterate: Find the two clusters that are closest to each other (based on linkage criteria) and merge them into a single cluster.

End: Repeat the process until only one single cluster (containing all data points) remains.

Q.20) What makes Silhouette Score a better metric than just inertia for model evaluation?

Answer ->

Inertia has a flaw: it always decreases as you increase $K$. If you set $K = N$ (number of points), inertia is 0, which is mathematically "perfect" but practically useless.

Silhouette Score is better because:

Bounded: It is normalized (-1 to 1), making it easier to interpret.

Separation & Cohesion: It measures not just how compact a cluster is (cohesion), but also how well-separated it is from other clusters.

Penalty for Overfitting: Unlike inertia, the Silhouette Score will decrease if you split a valid cluster into two unnecessary pieces, making it more effective for finding the true $K$.

Q.31) Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X, _ = make_blobs(n_samples=300, centers=3, cluster_std=2.0, random_state=42)
kmeans = KMeans(n_clusters=3).fit(X)

# Create meshgrid
h = .02
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Predict on grid
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot
plt.imshow(Z, interpolation='nearest', extent=(xx.min(), xx.max(), yy.min(), yy.max()),
           cmap=plt.cm.Pastel2, aspect='auto', origin='lower')
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, edgecolors='k')
plt.title("K-Means Decision Boundaries")
plt.show()
```

Q.44) Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import numpy as np

iris = load_iris()
X = iris.data
kmeans = KMeans(n_clusters=3, random_state=42).fit(X)
labels = kmeans.labels_
sample_vals = silhouette_samples(X, labels)

# Plotting sorted silhouette values
y_lower = 10
for i in range(3):
    ith_vals = sample_vals[labels == i]
    ith_vals.sort()
    size_cluster_i = ith_vals.shape[0]
    y_upper = y_lower + size_cluster_i
    plt.fill_betweenx(np.arange(y_lower, y_upper), 0, ith_vals, alpha=0.7)
    y_lower = y_upper + 10

plt.title("Silhouette Coefficient for Each Sample")
plt.xlabel("Coefficient Value")
plt.ylabel("Cluster")
plt.show()
```

Q.46) Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4 features)

Q.47) Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import collections

X, _ = make_blobs(n_samples=300, centers=3, cluster_std=2.0, random_state=42)

# Use strict eps to force some points to be noise
dbscan = DBSCAN(eps=1.0, min_samples=5)
labels = dbscan.fit_predict(X)

counts = collections.Counter(labels)
print("Label Counts (Label -1 is Noise):", counts)
```

Q.48) Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the clusters

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering

digits = load_digits()
X_embedded = TSNE(n_components=2, random_state=42).fit_transform(digits.data)

labels = AgglomerativeClustering(n_clusters=10).fit_predict(X_embedded)

plt.scatter(X_embedded[:, 0], X_embedded[:, 1], c=labels, cmap='nipy_spectral', s=10)
plt.title("Agglomerative Clustering on t-SNE Digits")
plt.show()
```
