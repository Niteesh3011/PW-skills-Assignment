## Question 1: Compare and contrast NLTK and spaCy in terms of features, ease of use,
and performance.
Answer ->

1. Features 

NLTK (The Toolbox): It is a massive collection of libraries and corpora. It provides multiple ways to do the same thing (e.g., several different stemmers like Porter, Lancaster, and Snowball). It is excellent for granular control and experimenting with specific linguistic theories.

spaCy (The Service): It is opinionated. Instead of giving you ten ways to tokenize, it gives you the one way the authors believe is most efficient. It features built-in support for deep learning integration, pre-trained word vectors, and highly accurate Named Entity Recognition (NER).

2. Ease of Use

NLTK: Can be cumbersome for beginners because it requires you to manually choose and chain together different modules. It primarily works with strings, meaning you often have to manage data structures yourself.

spaCy: Uses a "Document" object approach. When you process a text, spaCy automatically runs a pipeline (tokenization, tagging, parsing) and stores all the information in a single object. This makes it much more intuitive for developers who want to extract information quickly with minimal boilerplate code.

3. Performance

Speed: spaCy is significantly faster. It is written in Cython, which allows it to execute at C-like speeds. In benchmarks, spaCy's tokenizer and POS-tagger often outperform NLTK by a factor of 10x or more.

Memory: NLTK is generally more "lightweight" in terms of initial memory footprint because you only load what you need. spaCy requires loading pre-trained models into memory, which can be resource-intensive, though it is much more efficient at processing large batches of text once loaded.

# Question 4: Architecture and Functioning of a Recurrent Neural Network (RNN)

## 🔹 What is an RNN?

A **Recurrent Neural Network (RNN)** is a type of neural network designed to process **sequential data** such as:
- Text (sentences, documents)  
- Time series data  
- Speech signals  

Unlike traditional neural networks, RNNs **retain information from previous inputs** using a **hidden state (memory)**.

---

## 🔹 Architecture of an RNN

An RNN processes data **step-by-step over time (time steps)** and maintains a connection between previous and current inputs.

### 👉 Core Components:
- **Input (xₜ)** → current input at time step *t*  
- **Hidden State (hₜ)** → memory from previous steps  
- **Output (yₜ)** → output at time step *t*  

---

## 🔸 Mathematical Representation

### Hidden State Update:
\[
h_t = f(W_h h_{t-1} + W_x x_t + b)
\]

Where:
- \(h_t\): current hidden state  
- \(h_{t-1}\): previous hidden state  
- \(x_t\): current input  
- \(W_h, W_x\): weight matrices  
- \(b\): bias  
- \(f\): activation function (tanh/ReLU)  

### Output Equation:
\[
y_t = g(W_y h_t + c)
\]

---

## 🔹 Functioning of RNN (Step-by-Step)

1. Input sequence is fed one element at a time  
2. First input \(x_1\) generates hidden state \(h_1\)  
3. Hidden state \(h_1\) is passed to next step  
4. Next input \(x_2\) combines with \(h_1\) → produces \(h_2\)  
5. Process continues for entire sequence  
6. Output is generated at each step or final step  

👉 This creates a **loop-like (recurrent) structure**, enabling memory.

---

## 🔹 Key Idea: Memory

- RNN stores past information in **hidden states**  
- Helps in understanding **context and sequence dependencies**  

### 👉 Example:
> “I grew up in France… I speak fluent ___”  

RNN uses previous context → predicts **French**

---

## 🔹 Types of RNN Architectures

- **One-to-One** → Simple neural network  
- **One-to-Many** → Image captioning  
- **Many-to-One** → Sentiment analysis  
- **Many-to-Many** → Machine translation  

---

## 🔹 Advantages

- Handles sequential data efficiently  
- Maintains context using memory  
- Useful in:
  - Language modeling  
  - Translation  
  - Speech recognition  

---

## 🔹 Limitations

- **Vanishing gradient problem**  
- Difficulty in learning long-term dependencies  

---

## 🔹 Solutions

- **LSTM (Long Short-Term Memory)**  
- **GRU (Gated Recurrent Unit)**  

These improve RNN performance for long sequences.

---

## 🔹 Conclusion

RNN is a powerful neural network for **sequence modeling** that uses **hidden states as memory**. It processes data step-by-step and is widely used in NLP and time-series tasks, though improved models like LSTM and GRU are often preferred.

# Question 5: Difference between LSTM and GRU in NLP

## 🔹 Introduction

Both **LSTM (Long Short-Term Memory)** and **GRU (Gated Recurrent Unit)** are advanced types of **Recurrent Neural Networks (RNNs)**.

- They are designed to solve the **vanishing gradient problem**
- Help in learning **long-term dependencies** in sequential data
- Widely used in NLP tasks like translation, text generation, and sentiment analysis

---

## 🔹 LSTM (Long Short-Term Memory)

LSTM has a **more complex architecture** with three gates:

### 🔸 Gates in LSTM:
1. **Forget Gate** → decides what information to discard  
2. **Input Gate** → decides what new information to store  
3. **Output Gate** → decides what to output  

👉 It also maintains a **cell state (long-term memory)**

---

## 🔹 GRU (Gated Recurrent Unit)

GRU is a **simplified version of LSTM** with fewer gates:

### 🔸 Gates in GRU:
1. **Update Gate** → combines forget + input functions  
2. **Reset Gate** → controls how much past information to forget  

👉 Does NOT have a separate cell state (uses hidden state only)

---

## 🔹 Key Differences between LSTM and GRU

| Feature              | LSTM                          | GRU                          |
|---------------------|-------------------------------|------------------------------|
| Number of Gates     | 3 (Forget, Input, Output)     | 2 (Update, Reset)            |
| Cell State          | Present                       | Not present                  |
| Complexity          | More complex                  | Simpler                      |
| Training Speed      | Slower                        | Faster                       |
| Performance         | Better for long sequences     | Comparable, sometimes better |
| Parameters          | More                          | Fewer                        |

---

## 🔹 Advantages

### LSTM:
- Better for **long-term dependencies**
- More control over memory

### GRU:
- Faster to train  
- Requires less data  
- Simpler architecture  

---

## 🔹 When to Use What?

- Use **LSTM** when:
  - You need to capture **long sequences**
  - Accuracy is more important  

- Use **GRU** when:
  - You want **faster training**
  - Dataset is smaller  
  - Simpler model is sufficient  

---

## 🔹 Conclusion

LSTM and GRU are both powerful RNN variants.  
- **LSTM** is more powerful but complex  
- **GRU** is faster and simpler  

👉 Choice depends on the problem, dataset size, and performance requirements.

# Question 8: LSTM Model for Text Classification using Keras

## 🔹 Problem Statement

Build a **basic LSTM model** using Keras to classify sentences as:
- **Positive (1)**
- **Negative (0)**

---

## 🔹 Dataset

```python
texts = [
    "I love this project",            # Positive
    "This is an amazing experience",  # Positive
    "I hate waiting in line",         # Negative
    "This is the worst service",      # Negative
    "Absolutely fantastic!"           # Positive
]

labels = [1, 1, 0, 0, 1]

````python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# Dataset
texts = [
    "I love this project",
    "This is an amazing experience",
    "I hate waiting in line",
    "This is the worst service",
    "Absolutely fantastic!"
]

labels = np.array([1, 1, 0, 0, 1])

# 🔹 Tokenization
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)

sequences = tokenizer.texts_to_sequences(texts)

# 🔹 Padding
max_len = max(len(seq) for seq in sequences)
padded_sequences = pad_sequences(sequences, maxlen=max_len, padding='post')

# 🔹 Vocabulary size
vocab_size = len(tokenizer.word_index) + 1

# 🔹 Build LSTM Model
model = Sequential()

model.add(Embedding(input_dim=vocab_size, output_dim=8, input_length=max_len))
model.add(LSTM(16))
model.add(Dense(1, activation='sigmoid'))

# 🔹 Compile Model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# 🔹 Train Model
model.fit(padded_sequences, labels, epochs=10, verbose=1)

# 🔹 Model Summary
model.summary()
````

# Question 9: NLP Pipeline using spaCy

## Problem Statement
Using spaCy, process the paragraph below and print:
- each token
- its lemma
- all named entities

Dataset text:
"Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the development of India's atomic energy program. He was the founding director of the Tata Institute of Fundamental Research (TIFR) and was instrumental in establishing the Atomic Energy Commission of India."

````python
import spacy

# Load English model
nlp = spacy.load("en_core_web_sm")

# Given paragraph
text = """Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role 
in the development of India’s atomic energy program. He was the founding director 
of the Tata Institute of Fundamental Research (TIFR) and was instrumental in 
establishing the Atomic Energy Commission of India."""

# Process text
doc = nlp(text)

# 🔹 Tokenization and Lemmatization
print("Tokens and Lemmas:\n")
for token in doc:
    print(f"{token.text} -> {token.lemma_}")

# 🔹 Named Entity Recognition
print("\nNamed Entities:\n")
for ent in doc.ents:
    print(f"{ent.text} ({ent.label_})")
````

Design a chatbot for a **mental health platform** using:

## 🔹 Data Preprocessing Pipeline

1. **Text Cleaning**
   - Lowercasing  
   - Removing punctuation  

2. **spaCy Processing**
   - Tokenization  
   - Lemmatization  

3. **Tokenization (Keras)**
   - Convert words → sequences  

4. **Padding**
   - Make all sequences equal length  

---

## 🔹 Python Implementation (Simplified)

```python
import spacy
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, GRU, Dense

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Sample dataset
texts = [
    "I feel very sad",
    "I am happy today",
    "I am anxious about exams",
    "I feel relaxed"
]

labels = np.array([0, 1, 0, 1])  # 0 = negative, 1 = positive

# 🔹 Preprocessing using spaCy
def preprocess(text):
    doc = nlp(text.lower())
    return " ".join([token.lemma_ for token in doc if token.is_alpha])

clean_texts = [preprocess(t) for t in texts]

# 🔹 Tokenization
tokenizer = Tokenizer()
tokenizer.fit_on_texts(clean_texts)

sequences = tokenizer.texts_to_sequences(clean_texts)
padded = pad_sequences(sequences, padding='post')

# 🔹 Model (GRU)
model = Sequential()
model.add(Embedding(input_dim=len(tokenizer.word_index)+1, output_dim=8))
model.add(GRU(16))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# 🔹 Training
model.fit(padded, labels, epochs=10, verbose=1)

# 🔹 Test Prediction
test_text = ["I feel stressed"]
test_clean = [preprocess(t) for t in test_text]
test_seq = tokenizer.texts_to_sequences(test_clean)
test_pad = pad_sequences(test_seq, maxlen=padded.shape[1])

prediction = model.predict(test_pad)

print("Prediction:", prediction)
