#### Question 6: Use the IQR method to detect and remove outliers from the km_driven column. Show before-and-after summary statistics.


```python
sns.boxplot(y=df['km_driven'],x = df['owner'],palette='Set2')
```

```python
IQR = df['km_driven'].quantile(0.75) - df['km_driven'].quantile(0.25)
lower_bound = df['km_driven'].quantile(0.25) - (IQR * 1.5)
upper_bound = df['km_driven'].quantile(0.75) + (IQR * 1.5)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Before removing outliers:")
print(df['km_driven'].describe())
df = df[(df['km_driven'] >= lower_bound) & (df['km_driven'] <= upper_bound)]
print("After removing outliers:")
print(df['km_driven'].describe())
```
